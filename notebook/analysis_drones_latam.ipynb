# Drone Weaponization by Non-State Armed Groups in Latin America (2021–2026)

On October 28, 2025, security forces of the State of Rio de Janeiro launched “Operação Contenção” to contain the advance of the criminal faction Comando Vermelho (CV). Approximately 2,500 agents took part in confrontations across 26 communities in the city’s North Zone, primarily in Complexo da Penha and Complexo do Alemão. The operation drew significant national and international media attention — first, for its death toll: over 121 people, including 4 police officers, according to official data from the Civil and Military Police, making it the deadliest in the country’s history. Second, for the unprecedented and extensive use of drones against state agents.

Prior to this, drone use by criminals in Brazil had been predominantly associated with surveillance, drug trafficking, and inter-faction disputes. Researcher Roberto Uchôa argues that the militarization of drone use by organized crime has documented origins in the conflict in Ukraine, spreading through trafficking routes across Latin America due to its accessibility and low cost. Mostly manufactured in China, the drones are cheap and widely available. The Eastern European battlefield functioned as a laboratory for tactical innovation and contributed to the diffusion of these practices among non-state armed actors, pioneered in the region by Mexican cartels such as the CJNG and Sinaloa. According to The Economist, thousands of Colombian and Mexican fighters are said to have enlisted in the Ukraine war to gain experience with drones. 

The first recorded use of unmanned aerial systems (UAS) by non-state armed groups for monitoring occurred in 2006, in southern Lebanon, during the 2006 Lebanon War. Today, this technology is being deployed by organized crime in Mexico, Brazil, and Colombia as vectors of both transport and attack, which translate into a new security framework in Latin America. It transformed transnational operations and local trafficking networks, breaking the state’s aerial monopoly and sophisticating criminal logistics, territorial control, and confrontation tactics. Once again, organized crime has proven highly adaptable, contributing to the growing trend of asymmetric conflict in recent decades.

## Data and Methodology

The 30 incidents were manually collected from open sources, including Gazeta do Povo, BBC, El Tiempo, and InSight Crime. The data covers Brazil, Mexico, and Colombia from 2021 through the first half of 2026 and represents a sample, not a comprehensive record of incidents. This project uses Python (pandas, matplotlib, folium, geopy/Nominatim) for visualization and analysis.

```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
import folium

PROJECT_DIR = Path.cwd()
DATA_PATH = PROJECT_DIR / "incidentes_drones.xlsx"

df = pd.read_excel(DATA_PATH)
df = df.dropna(how="all")

df["data"] = pd.to_datetime(df["data"])
df["ano"] = df["data"].dt.year

df["alvo_agrupado"] = df["alvo"].replace({
    "policia": "Forças de segurança",
    "exercito": "Forças de segurança",
    "forças de segurança": "Forças de segurança",
    "civis": "População civil",
    "expansão territorial": "Facção rival",
    "massa carcerária": "Facção rival",
    "comando vermelho": "Facção rival",
    "los tlacos": "Facção rival",
    "cartel de sinaloa": "Facção rival",
    "carteles unidos": "Facção rival",
    "milícia": "Facção rival",
    "infraestrutura governamental": "Infraestrutura",
})

print(f"Dataset loaded: {len(df)} incidents")
```

```python
required_columns = {
    "data",
    "pais",
    "estado_regiao",
    "latitude",
    "longitude",
    "grupo",
    "alvo",
    "descricao",
    "fonte",
    "url",
}

missing_columns = required_columns - set(df.columns)

if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

print("All required columns are present.")
```

Chart 1 — Targets

Based on available data, drone use by organized crime remains predominantly associated with surveillance and drug trafficking. This analysis focuses specifically on weaponized attacks and their targets. In this context, security forces account for the majority of documented incidents, followed by rival factions, indicating that drones are being deployed as instruments of territorial control and armed confrontation. This is consistent with a broader shift in which organized crime has increasingly prioritized community domination over direct drug sales, a trend particularly evident in Rio de Janeiro, where territorial control has become more profitable and diversified than the drug trade itself.

Chart 3 — By Year and Country

Mexican cartel pioneership reflects several factors. The CJNG and Sinaloa operate on the border with the world’s largest commercial drone market, the United States, which facilitates access to technology, components, and technical knowledge through the same channels used to move drugs and weapons. The CJNG, in particular, has a history of accelerated militarization, demonstrating the capacity to invest in tactical innovation systematically. Mexico has also been engaged in an intense inter-cartel war for over a decade, an environment that creates significant pressure for innovation.

Similarly, Colombian armed groups have embraced these innovations as part of broader strategies of expansion and territorial consolidation. The failure of Gustavo Petro’s paz total policy created the conditions for this escalation. Criminal groups exploited ceasefire periods to regroup, rearm, and train, before resuming operations with greater tactical sophistication, particularly in Cauca and Catatumbo.

In Brazil, the CV and PCC (Primeiro Comando da Capital) expanded operations nationally, bringing conflict to states previously peripheral to structured organized crime. In 2024 alone, the Federal Police seized R$5.6 billion in faction assets, a 70% increase over the previous year, reflecting growing militarization and financial sophistication. The country also rose in global rankings for deaths in armed confrontations between 2024 and 2025, with Bahia, Ceará, and Rio de Janeiro as epicenters.

These findings show an accelerating trend that reflects a new form of asymmetric warfare between organized crime and the state. Driven by expanding territorial ambitions and diversifying revenue streams, non-state armed groups have demonstrated a growing capacity for tactical sophistication. While they do not exercise absolute dominance, the pressure they exert on state institutions is significant and demands urgent, coordinated responses, including investment in UAS import regulation and traceability, development of technical countermeasures, and the protection of affected populations. Future research could investigate procurement and supply chain networks, including the role of commercial platforms, intermediaries, and existing smuggling routes, to better establish how drone technology reaches these groups and whether diffusion follows predictable pathways that could inform interdiction efforts.

## Optional step: geocode missing coordinates

This section only searches for coordinates when new incidents are missing latitude or longitude. If all rows already have coordinates, the code will skip geocoding automatically.

```python
from geopy.geocoders import Nominatim
import time

missing_coordinates = df["latitude"].isna() | df["longitude"].isna()

if missing_coordinates.any():
    geolocator = Nominatim(user_agent="projeto_drones")

    def buscar_coordenadas(row):
        local = f"{row['estado_regiao']}, {row['pais']}"
        try:
            location = geolocator.geocode(local)
            time.sleep(1)
            if location:
                return location.latitude, location.longitude
        except Exception:
            pass
        return None, None

    df.loc[missing_coordinates, ["latitude", "longitude"]] = df.loc[
        missing_coordinates
    ].apply(buscar_coordenadas, axis=1, result_type="expand")

    df.to_excel(DATA_PATH, index=False)

    print(f"Coordinates saved: {df['latitude'].notna().sum()} of {len(df)}")
else:
    print("All rows already have coordinates. Geocoding was skipped.")
```

```python
cores_alvo = {
    'Forças de segurança': 'red',
    'Facção rival': 'orange',
    'População civil': 'yellow',
    'Infraestrutura': 'purple'
}

traducao_alvo = {
    'Forças de segurança': 'Security forces',
    'Facção rival': 'Rival faction',
    'População civil': 'Civilian population',
    'Infraestrutura': 'Infrastructure'
}

# ---- MAPA PT ----
mapa_pt = folium.Map(location=[-5, -70], zoom_start=4)

for _, row in df.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        cor = cores_alvo.get(row['alvo_agrupado'], 'gray')
        popup_texto = f"""
        <b>Grupo:</b> {str(row['grupo']).replace('"', '')}<br>
        <b>País:</b> {str(row['pais'])}<br>
        <b>Local:</b> {str(row['estado_regiao'])}<br>
        <b>Ano:</b> {str(row['data'])[:4]}<br>
        <b>Alvo:</b> {str(row['alvo_agrupado'])}<br>
        <b>Descrição:</b> {str(row['descricao']).replace('"', '')}<br>
        <b>Fonte:</b> <a href="{str(row['url'])}" target="_blank">{str(row['fonte'])}</a>
        """
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=7, color=cor, fill=True,
            fill_color=cor, fill_opacity=0.8,
            popup=folium.Popup(popup_texto, max_width=300)
        ).add_to(mapa_pt)

legenda_pt = """
<div style="position: fixed; bottom: 40px; left: 40px; z-index: 1000;
     background-color: white; padding: 12px; border-radius: 8px;
     border: 1px solid grey; font-size: 13px;">
    <b>Categoria de alvo</b><br>
    <span style="color:red;">●</span> Forças de segurança<br>
    <span style="color:orange;">●</span> Facção rival<br>
    <span style="color:goldenrod;">●</span> População civil<br>
    <span style="color:purple;">●</span> Infraestrutura
</div>
"""
mapa_pt.get_root().html.add_child(folium.Element(legenda_pt))
mapa_pt.save('mapa_drones_pt.html')

# ---- MAPA EN ----
mapa_en = folium.Map(location=[-5, -70], zoom_start=4)

for _, row in df.iterrows():
    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        alvo_en = traducao_alvo.get(row['alvo_agrupado'], row['alvo_agrupado'])
        cor = cores_alvo.get(row['alvo_agrupado'], 'gray')
        popup_texto = f"""
        <b>Group:</b> {str(row['grupo']).replace('"', '')}<br>
        <b>Country:</b> {str(row['pais'])}<br>
        <b>Location:</b> {str(row['estado_regiao'])}<br>
        <b>Year:</b> {str(row['data'])[:4]}<br>
        <b>Target:</b> {alvo_en}<br>
        <b>Description:</b> {str(row['descricao']).replace('"', '')}<br>
        <b>Source:</b> <a href="{str(row['url'])}" target="_blank">{str(row['fonte'])}</a>
        """
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=7, color=cor, fill=True,
            fill_color=cor, fill_opacity=0.8,
            popup=folium.Popup(popup_texto, max_width=300)
        ).add_to(mapa_en)

legenda_en = """
<div style="position: fixed; bottom: 40px; left: 40px; z-index: 1000;
     background-color: white; padding: 12px; border-radius: 8px;
     border: 1px solid grey; font-size: 13px;">
    <b>Target category</b><br>
    <span style="color:red;">●</span> Security forces<br>
    <span style="color:orange;">●</span> Rival faction<br>
    <span style="color:goldenrod;">●</span> Civilian population<br>
    <span style="color:purple;">●</span> Infrastructure
</div>
"""
mapa_en.get_root().html.add_child(folium.Element(legenda_en))
mapa_en.save('mapa_drones_en.html')

print("Mapas salvos!")
```
